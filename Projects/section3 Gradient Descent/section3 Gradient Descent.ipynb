### Notebook import and Packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

### Example 1 cost function
$$f(x) = x^2 + x + 1$$

```python
def f(x):
    return x**2 + x + 1
```

```python
# Making the data
x_1 = np.linspace(start=-3, stop=3, num=100)
```

```python
plt.plot(x_1, f(x_1))
plt.ylim(0,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.show()
```

### Derivatives and Slope

```python
def df(x):
    return 2*x + 1
```

```python
# Plotting function and its derivatve 
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.plot(x_1, f(x_1), color='skyblue', linewidth=3)
plt.title("Cost Function", fontsize=17)
plt.ylim(0,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("f(x)", fontsize=16)

# Derivative
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color='blue', linewidth=3)
plt.title("Derivative Cost Function", fontsize=17)
plt.grid()
plt.ylim(-6,6)
plt.xlabel("x", fontsize=16)
plt.ylabel("df(x)", fontsize=16)

plt.show()
```

```python
# Gradient Descent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001
x_list = []
slope_list = []

while abs(new_x - previous_x) > precision:
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    x_list.append(previous_x)
    slope_list.append(df(previous_x))
    
print('Local minimum at: ', new_x)
print('Slope at the point is: ', gradient)
print('f(x) value of cost function at this point is: ', f(new_x))
```

```python
# Superimposing the data Plotting function and its derivatve 
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.plot(x_1, f(x_1), color='skyblue', linewidth=3)
plt.title("Cost Function", fontsize=17)
plt.ylim(0,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("f(x)", fontsize=16)

# Adding the gradient descent data (f(array) works but not f(list))
values = np.array(x_list) 
plt.scatter(x_list, f(values), color='blue', s=50, alpha=0.5)

# Derivative
plt.subplot(1, 2, 2)
plt.plot(x_1, df(x_1), color='blue', linewidth=3, alpha=0.5)
plt.title("Derivative Cost Function", fontsize=17)
plt.grid()
plt.ylim(-6,6)
plt.xlabel("x", fontsize=16)
plt.ylabel("df(x)", fontsize=16)

# Adding the gradient descent data (plt only works with arrays)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()
```

# Example 2 - multiple minima vs Initial guess & Advanced function
$$g(x) = x^4 - 4x^2 + 5$$

```python
def g(x):
    return x**4 - 4*x**2 + 5
def dg(x):
    return 4*x**3 - 8*x

# Making the data
x_2 = np.linspace(start=-2, stop=2, num=1000)
```

```python
# Plotting function g 
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.plot(x_2, g(x_2), color='skyblue', linewidth=3)
plt.title("Cost Function", fontsize=17)
plt.ylim(0,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("g(x)", fontsize=16)

# Adding the gradient descent data (f(array) works but not f(list))
values = np.array(x_list) 
plt.scatter(x_list, f(values), color='blue', s=50, alpha=0.5)

# Plotting the dg
plt.subplot(1, 2, 2)
plt.plot(x_2, dg(x_2), color='blue', linewidth=3, alpha=0.5)
plt.title("Derivative Cost Function", fontsize=17)
plt.grid()
plt.ylim(-8,8)
plt.xlabel("x", fontsize=16)
plt.ylabel("dg(x)", fontsize=16)

# Adding the gradient descent data (plt only works with arrays)
plt.scatter(x_list, slope_list, color='red', s=100, alpha=0.5)

plt.show()
```

```python
# Functionalizing gradient descent add ; to the function call to supress the input 
def gradient_descend(new_x: int, step_multipliter: int, precision, f, df):
    x_list = []
    slope_list = []
    previous_x = 0
    
    while abs(new_x - previous_x) > precision:
        previous_x = new_x
        gradient = df(previous_x)
        new_x = previous_x - step_multiplier * gradient

        x_list.append(previous_x)
        slope_list.append(df(previous_x))
        
    print('Local minimum at: ', new_x)
    print('Slope at the point is: ', gradient)
    print('f(x) value of cost function at this point is: ', f(new_x))
    
    return x_list, slope_list
```

```python
# Gradient Descent for g
new_x = -2
step_multiplier = 0.1
precision = 0.0001

gradient_descend(new_x, step_multiplier, precision, g, dg);

```
